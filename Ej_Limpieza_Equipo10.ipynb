##LIMPIEZA DE DATOS

**Leer el archivo**

EL primer paso sera importar las librerias las cuales podremos utilizar para la lectura y limpieza de datos. Despues se mandará llamar el archivo el cual esta en formato excel. Para eliminar los posibles simbolos que no se puedan leer utilizaremos encoding.

**Manejo de Columnas**

Para facilitar nuestra visualización de los datos nosotros decidiremos eliminar columnas las cuales no nos den una informacion relevante para nuestro objetivo. para decidir que columnas eliminaremos primero veremos el nombre de las columnas que tenemos y despues el tipo de dato que contienen


```python
google.columns
```

```python
google.dtypes
```

Podemos ver que existen diferentes columnas las cuales no nos serviran para nuestro objetivo y el tenerlas mostradas solo entorpecera la visualizacion de nuestros datos, por lo cual las eliminaremos.


```python
google1=google.drop(['Size','Price','Genres','Last Updated','Current Ver','Android Ver'],axis=1)
google1
```

Ahora cambiaremos el nombre de las columnas para que nos sea mas facil identificar de que esta hablando cada columna.


```python
google1.rename(columns={'App':'Nombre de la Aplicacion'}, inplace = True)
google1.rename(columns={'Category':'Categoria'},inplace = True)
google1.rename(columns={'Rating':'Calificacion'},inplace = True)
google1.rename(columns={'Reviews':'Numero de reseñas'},inplace = True)
google1.rename(columns={'Installs':'Instaladas'},inplace = True)
google1.rename(columns={'Type':'Gratis/Pago'},inplace = True)
google1.rename(columns={'Content Rating':'Clasificacion'},inplace = True)

google1.head()
```

Revisaremos si existen columnas con datos nulos.

```python
google1.isnull().any()
```

```python
google1.isnull().sum()
```

Dado que si existen valores nulos en nuestras filas y estos datos nos apoyaran a llegar a nuestro obetivo los eliminaremos.

```python
g2=google1.dropna()
```

Revisamos si funcionó

```python
g2.isnull().any()
```

Ordenaremos las columnas de forma que los datos mas relevantes, para nosotros, se presenten primero.

```python
#https://www.delftstack.com/es/howto/python-pandas/how-to-change-the-order-of-dataframe-columns/
#https://www.delftstack.com/es/howto/python-pandas/pandas-remove-index/#:~:text=Si%20queremos%20eliminar%20la%20columna,en%20el%20m%C3%A9todo%20reset_index()%20.
g3=g2.reindex(columns=['Nombre de la Aplicacion','Calificacion','Numero de reseñas','Instaladas','Categoria','Clasificacion','Gratis/Pago'])
g3=g3.set_index('Nombre de la Aplicacion')
g3
```

Ahora tomaremos solo las calificaciones que tienen una buena calificacion (mayor a 3).

```python
g4=g3[g3['Calificacion'] > 3]
g4
```

Por ultimo ordenaremos los datos de la calificacion de manera descendente para asi tener las aplicaciones con mejor puntuacion primero.

```python
#https://www.analyticslane.com/2019/04/29/diferentes-formas-de-ordenar-dataframes-en-pandas/

google_final=g4.sort_values('Calificacion',ascending=False)
```

```python
google_final.head()
```

```python
google_final.tail()
```

Guardaremos nuestro nuevo data frame en un archivo nuevo.
