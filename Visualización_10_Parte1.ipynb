HEATMAP

```python
import matplotlib.pyplot as plt
```

```python
f,ax=plt.subplots(figsize = (18,18))
sns.heatmap(df_ga.corr(),annot= True,linewidths=0.5,fmt = ".1f",ax=ax)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.title('Mapa de correlación')
plt.savefig('graph.png')
plt.show()
```

Scatterplot heatmap

```python
corr_mat = df_ga.corr().stack().reset_index(name="correlation")
g = sns.relplot(
    data=corr_mat,
    x="level_0", y="level_1", hue="correlation", size="correlation",
    palette="vlag", hue_norm=(-1, 1), edgecolor=".7",
    height=10, sizes=(50, 250), size_norm=(-.2, .8),
)
g.set(xlabel="", ylabel="", aspect="equal")
g.despine(left=True, bottom=True)
g.ax.margins(.02)
for label in g.ax.get_xticklabels():
    label.set_rotation(90)
for artist in g.legend.legendHandles:
    artist.set_edgecolor(".7")
```

El scatterplot heatmap y el heatmap parecen dar la misma información que es la correlación de los datos mas sin embargo hay una diferencia en la presentación de estos datos podemos observar que el heatmap te pone toda la información en cuadros con el respectivo valor de la correlación mientras que en el scatterplot lo tenemos con circulos mas chicos a simple vista es mas digerible porque la escala de colores te la pone a un lado y no satyra el gráfico de información. Por lo que el mejor gráfico en lo personal seria el scatterplot heatmap
